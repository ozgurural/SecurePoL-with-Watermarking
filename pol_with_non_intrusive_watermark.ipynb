```python
# non-intrusive PoL + watermark  (λ = 0.03, k = 1, 100 epochs)
!cd /content/drive/MyDrive/SecurePoL-with-Watermarking/PoL && \
python train.py \
  --dataset CIFAR10 \
  --model resnet20 \
  --epochs 100 \
  --save-freq 100 \
  --lr 0.1 \
  --lambda-wm 0.03 \
  --k 1 \
  --watermark-method non_intrusive \
  --watermark-key secret_key \
  --watermark-size 128 \
  --tolerance-wm 1 \
  --log-tb \
  --log-dir  /content/drive/MyDrive/SecurePoL-with-Watermarking/logs/non_intrusive \
  --model-dir /content/drive/MyDrive/SecurePoL-with-Watermarking/proof/CIFAR10_non_intrusive
```
